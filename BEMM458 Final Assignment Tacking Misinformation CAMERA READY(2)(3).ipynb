## Question 3. Compute the z-score of the column "public_metrics.retweet_count" (15 points)
## tip: refer to the definition of Z-score at" https://www.statisticshowto.com/probability-and-statistics/z-score/

```python
#Null values
spainEnglishAll.dropna(inplace=True)
```

```python
#Creating a function to calculate the Z-score of the "public_metrics.retweet columns"
def zscore(df):
    Mean = df["public_metrics.retweet_count"].mean()
    standard_dev = df["public_metrics.retweet_count"].std()
    Zscore = (df["public_metrics.retweet_count"]-Mean)/float(standard_dev)
    df["zscore"]= Zscore
    return df
```

```python
dfnew = zscore(spainEnglishAll)
```

```python
dfnew.head(5)
```

## Question 4. Filter conversations with a z-score larger than 10. (15 points)

## Question 6. Extract, from the text colum, information on who created the conversation in the first place (15 points)
### Tip: retweeted conversations start with the pattern RT@
### For illustration purposes, the text "RT @20splentyforus: France, like UK should set a national 30km/h or 20mph urban default limit." was originally created by @20splentyforus
### Tip2. consider applying a regex pattern to the DataFrame using apply method

```python
import re
```

```python
#filtering all the tweets starting with RT
filteredRTtweets = spainEnglishAll[spainEnglishAll["tweet.text"].apply(lambda x:bool(re.search("^RT",x)))]
#getting all the names of people who created tweet originally
filteredRTtweets["Original Creator"] = filteredRTtweets["tweet.text"].apply(lambda x:re.findall("(?<=RT )[^:]+",x))


```
